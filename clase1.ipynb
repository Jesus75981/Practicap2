```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('Wey.jpg',0)
mascara  = np.array([[ 0, -1,  0], [-1, 5, -1], [0, -1, 0]])
mascara2 = np.array([[1, 1, 1],[1, 1, 1],[1, 1, 1]]) / 9
mascara3 = np.array([[0, 1, 0], [ 1, -4,  1], [0, 1, 0]])
mascara4 = np.array([[0, 0, 0], [ -1, 1,  0], [0, 0, 0]])
mascara5 = np.array([[-2, -1, 0], [ -1, 1,  1], [0, 1, 2]])

imgFiltro1 = cv2.filter2D(img, -1, mascara)
imgFiltro2 = cv2.filter2D(img, -1, mascara2)
imgFiltro3 = cv2.filter2D(img, -1, mascara3)
imgFiltro4 = cv2.filter2D(img, -1, mascara4)
imgFiltro5 = cv2.filter2D(img, -1, mascara5)

plt.figure(figsize=(15, 8))
plt.subplot(321)
plt.title('Original')
plt.imshow(img ,cmap='gray')
plt.axis('off')

plt.subplot(322)
plt.title('Filtro 1')
plt.imshow(imgFiltro1 ,cmap='gray')
plt.axis('off')

plt.subplot(323)
plt.title('Filtro 2 (Difuminado)')
plt.imshow(imgFiltro2 ,cmap='gray')
plt.axis('off')

plt.subplot(324)
plt.title('Filtro 3 (Deteccion de bordes)')
plt.imshow(imgFiltro3 ,cmap='gray')
plt.axis('off')

plt.subplot(325)
plt.title('Filtro 4 (Realce de bordes)')
plt.imshow(imgFiltro4 ,cmap='gray')
plt.axis('off')

plt.subplot(326)
plt.title('Filtro 5 (Repuje)')
plt.imshow(imgFiltro5 ,cmap='gray')
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


imagen = cv2.imread('Ejemplo1.png', cv2.IMREAD_GRAYSCALE)


imagen_reducida = cv2.resize(imagen, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_NEAREST)


plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(imagen, cmap='gray')
plt.title('Imagen Original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(imagen_reducida, cmap='gray')
plt.title('Imagen con Aliasing')
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


imagen = cv2.imread('Imagen2.jpg', cv2.IMREAD_GRAYSCALE)


patron1 = np.zeros_like(imagen)
patron2 = np.zeros_like(imagen)

for i in range(imagen.shape[0]):
    for j in range(imagen.shape[1]):
        patron1[i, j] = (i + j) % 255 
        patron2[i, j] = (i - j) % 255  


patron_moire = (patron1 + patron2) % 255


plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(imagen, cmap='gray')
plt.title('Imagen Original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(patron_moire, cmap='gray')
plt.title('Patrón de Moiré')
plt.axis('off')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

imagen = cv2.imread('Ejemplo1.png')


imagen_suavizada = cv2.GaussianBlur(imagen, (5, 5), 0)


imagen_reducida = cv2.resize(imagen_suavizada, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_AREA)


plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB))
plt.title('Imagen Original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(imagen_reducida, cv2.COLOR_BGR2RGB))
plt.title('Imagen Reducida con Suavizado (Menor Aliasing)')
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


imagen = cv2.imread('Imagen2.jpg', cv2.IMREAD_GRAYSCALE)


kernel = np.ones((5, 5), np.float32) / 25
imagen_filtrada = cv2.filter2D(imagen, -1, kernel)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(imagen, cmap='gray')
plt.title('Imagen Original (Moiré)')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(imagen_filtrada, cmap='gray')
plt.title('Imagen Filtrada (Reducción de Moiré)')
plt.axis('off')

plt.show()
```
